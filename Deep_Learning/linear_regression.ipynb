```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Generate 200 x values between -10 and 10
x = np.linspace(-10, 10, 200)

# Generate y values: y = 2x + 3 + noise
true_slope = 2
true_intercept = 3
noise = np.random.normal(0, 1.5, 200)  # Gaussian noise with mean 0, std 1.5
y = true_slope * x + true_intercept + noise

# Create a DataFrame to store the data
data = pd.DataFrame({'x': x, 'y': y})

# Save the dataset to a CSV file
data.to_csv('linear_regression_data.csv', index=False)

# Plot the data points and the true line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.5, label='Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Synthetic Linear Regression Dataset (200 Points)')
plt.legend()
plt.grid(True)
plt.show()

# Display the first few rows of the dataset
print(data.head())
```

```python
import numpy as np

N = 200
x_plot = x
y_plot = y
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

x = np.hstack((np.ones((N, 1)), x))

w = np.array([0.,1.]).reshape(-1,1)

numOfIteration = 10000
cost = np.zeros((numOfIteration,1))
learning_rate = 0.000001
for i in range(1, numOfIteration):
    r = np.dot(x, w) - y
    cost[i] = 0.5*np.sum(r*r)
    w[0] -= learning_rate*np.sum(r)
    # correct the shape dimension
    w[1] -= learning_rate*np.sum(np.multiply(r, x[:,1].reshape(-1,1)))
    print(cost[i])
predict = np.dot(x, w)
plt.figure(figsize=(10, 6))
plt.scatter(x_plot, y_plot, color='blue', alpha=0.5, label='Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Synthetic Linear Regression Dataset (200 Points)')
plt.plot((x[0][1], x[N-1][1]),(predict[0], predict[N-1]), 'r')
plt.show()

x1 = 2.5
y1 = w[0] + w[1] * 2.5
print('The value of y when x=2.5 : ', y1)
```
